```python
# Import Dependencies
import matplotlib.pyplot as plt
import numpy as np
```

```python

time = np.arange(0, 130, 10)

danger_drop_speeds = [9, 8, 90, 85, 80, 70, 70, 65, 55, 60, 70, 65, 50]
railgun_speeds = [75, 70, 60, 65, 60, 45, 55, 50, 40, 40, 35, 35, 30]
```

```python

danger_drop, = plt.plot(time, danger_drop_speeds, marker = "*", label="Danger Drop")
railgun, = plt.plot(time, railgun_speeds, marker = "8", label="Railgun")
```

```python
plt.title("Danger Drop speed vs. Railgun speed over time")
plt.xlabel("Time(seconds)")
plt.ylabel("Speed(mph)")
```

```python
plt.xlim(0, 120)
plt.ylim(5, 95)
```

```python
plt.legend(handles=[danger_drop, railgun], loc="best")
```

```python
plt.grid()
```

```python
plt.show()
```
