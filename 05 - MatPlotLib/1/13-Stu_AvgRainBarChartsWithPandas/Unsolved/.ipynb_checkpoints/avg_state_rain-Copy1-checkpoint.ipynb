```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Load in csv

rain_df = pd.read_csv("../Resources/avg_rain_state.csv")
rain_df.head()
```

```python
# Set x axis and tick locations
x_axis = np.arange(len(rain_df))
tick_locations = [value for value in x_axis]
```

```python

plt.figure(figsize=(7, 20))
x_axis = np.arange(len(rain_df))
plt.barh(x_axis, rain_df["Inches"], color='green', alpha=0.5, align="center")
plt.yticks(tick_locations, rain_df["State"], rotation="horizontal")
```

```python
plt.ylim(-0.5, len(x_axis))
plt.xlim(0, max(rain_df["Inches"]+ 10))
```

```python
plt.title("Average Rainfall")
plt.ylabel("State")
plt.xlabel("Rainfall (Inches)")
```

```python
plt.tight_layout()
```
