```python
# Dependencies
import pandas as pd
import numpy as np
```

```python
# Save file path to variable
pokemon_csv = "Resources/Pokemon.csv"
```

```python
# Read with Pandas
pokemon_df = pd.read_csv(pokemon_csv)
pokemon_df.head()
```

```python
# Extract the following columns: "Type 1", "HP", "Attack", "Sp. Atk", "Sp. Def", and "Speed"
pokemon_type_df = pokemon_df[["Type 1", "HP", "Attack", "Sp. Atk", "Sp. Def", "Speed"]]
pokemon_type_df.head()
```

```python
# Create a dataframe of the average stats for each type of pokemon.
pokemon_group_df = pokemon_type_df.groupby(["Type 1"])

pokemon_comparison_df = pokemon_group_df.mean()
pokemon_comparison_df
```

```python
# Calculate the total power level of each type of pokemon by summing all of the stats together.
# Place the results into a new column.

pokemon_comparison_df["Total Stats"] = pokemon_comparison_df.sum(axis = 1) 

# Hint: Research the sum method from pandas.
pokemon_comparison_df["Total Stats"]
```

```python
pokemon_comparison_df
```

```python
# Bonus: Sort the table by strongest type and export the resulting table to a new CSV.
strongest_pokemon_df = pokemon_comparison_df.sort_values(["Total Stats"], ascending = False)
strongest_pokemon_df
```
