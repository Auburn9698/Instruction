```python
# Dependencies
from bs4 import BeautifulSoup
import requests
import pymongo
```

```python
url = 'https://www.theepochtimes.com/the-french-revolution-is-attacking-the-american-revolution_3723968.html?fbclid=IwAR1ZUD__j7nd-Au8Wex0q3NbuF2B44ZgpiJrwXg_NUg8kGMn-CINRkeeCHk'

# Retrieve page with the requests module
response = requests.get(url)
# Create BeautifulSoup object; parse with 'lxml'
soup = BeautifulSoup(response.text, 'lxml')
```

```python
soup
```

```python
# Examine the results and look for a div withe the class 'top-matter'
results = soup.find_all('p')
```

```python
print(results[0].prettify())
```

```python
print(results[1].prettify())
```
