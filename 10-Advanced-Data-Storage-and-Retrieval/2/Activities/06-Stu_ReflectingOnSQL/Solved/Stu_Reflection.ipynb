```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine
```

```python
# Create engine using the `demographics.sqlite` database file
engine = create_engine("sqlite:///../Resources/demographics.sqlite")
```

```python
# Declare a Base using `automap_base()`
Base = automap_base()
```

```python
# Use the Base class to reflect the database tables
Base.prepare(engine, reflect=True)
```

```python
# Print all of the classes mapped to the Base
Base.classes.keys()
```

```python
# Do shift-tab-tab here to bring up a menu that lets you view the columns.

Demographics.
```

```python
# Assign the demographics class to a variable called `Demographics`
Demographics = Base.classes.demographics
```

```python
# Create a session
session = Session(engine)
```

```python
# Use the session to query Demographics table and display the first 5 locations
for row in session.query(Demographics, Demographics.location).limit(5):
    print(row)
```

```python
# Use the session to query Demographics table and display the first 5 locations
for row in session.query(Demographics.id, Demographics.location).limit(5):
    print(row)
```

```python
# Use the session to query Demographics table and display all locations
for row in session.query(Demographics.id, Demographics.location):
    print(row)
```

```python
# BONUS: Query and print the number of unique Locations
# Hints: Look into counting and grouping operations in SQLAlchemy
locations = session.query(Demographics).group_by(Demographics.location).count()
print(locations)
```
