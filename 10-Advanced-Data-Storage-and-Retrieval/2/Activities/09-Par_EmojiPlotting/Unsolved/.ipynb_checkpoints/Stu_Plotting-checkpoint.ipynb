```python
# Import SQLAlchemy `automap` and other dependencies here
# YOUR CODE HERE
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, inspect
```

```python
# Create an engine for the `emoji.sqlite` database
engine = create_engine("sqlite:///../Resources/emoji.sqlite", echo=False)
```

## Reflect database and Query

```python
# Reflect Database into ORM class
Base = automap_base()
Base.prepare(engine, reflect=True)  # Prepare gets the engine passed in / reflect=True gives you back all the classes.

# Building a table object called Emoji:
Emoji = Base.classes.emoji  # This just makes it easer to read, instead of typing/reading Base.classes.emoji in every spot
#                             two cells below where "Emoji" is.
```

```python
# Start a session to query the database
session = Session(engine)
```

Use Matplotlib to create a horizontal bar chart and plot the emoji `score` in descending order. Use `emoji_char` as the y-axis labels. Plot only the top 10 emojis ranked by score

```python
# Query Emojis for `emoji_char`, `emoji_id`, and `score` and save the query into results
# YOUR CODE HERE

# SELECT emoji_char, emoji_id, score FROM emoji ORDER BY score DESC

results = session.query(Emoji.emoji_char, Emoji.emoji_id, Emoji.score).\
    order_by(Emoji.score.desc()).all()

results
```

```python
# Unpack the `emoji_id` and `scores` from results and save into separate lists
# YOUR CODE HERE

emoji_id = [result[1] for result in results[:10]]

scores = [result[2] for result in results[:10]]
```

```python
emoji_id
```

```python
scores
```

## Plot using Matplotlib

```python
# Create a horizontal bar chart and plot the `emoji_id` on the y-axis and the `score` on the x-axis
# Challenge: Try to plot the scores in descending order on the graph (The largest score is at the top)
# The ypos is setting up the number of bars on the y axis (a range of 1, 11), or a range from
# 1 to the length of scores + 1)

fig, ax = plt.subplots()
ypos = range(1, len(scores)+1)
ax.barh(ypos, scores[::-1])
ax.set_xlabel("score")
ax.set_ylabel("emoji")
ax.set_yticks(ypos)
ax.set_yticklabels(emoji_id[::-1])
ax.set_title("Emoji Scores")
fig.tight_layout()
plt.show()
```

## Plot using Pandas Plotting

Load the results into a Pandas DataFrame

```python
# Class addition:

results[:10]
```

```python
# Load the results into a pandas dataframe. Set the index to the `emoji_id`
# Try this with the middle line commented out, to see it with an index, but without it being the emoji_id

df = pd.DataFrame(results[:10], columns = ["emoji_char", "emoji_id", "emoji_score"])
df.set_index('emoji_id', inplace = True)
df.head(10)
```

Plot using Pandas

```python
# Plot the dataframe as a horizontal bar chart using pandas plotting
# The ::-1 part puts everything in reverse order:

df.iloc[::-1].plot.barh(title="emoji ranking")
plt.tight_layout()
plt.show()

# Plotting it with pandas requires a lot less code
```

```python
# BONUS: Use Pandas `read_sql_query` to load a query statement directly into the DataFrame

stmt = session.query(Emoji).\
    order_by(Emoji.score.desc()).statement
df2 = pd.read_sql_query(stmt, session.bind)
df2.head(10)
```
